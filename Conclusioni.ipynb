```python
import pandas as pd
import numpy as np
import seaborn as sns

from typing import List
import os

import matplotlib as mpl
import matplotlib.pyplot as plt
```

Risultati Accuracy su testSet in base ad algoritmo utilizzato 

```python
DT_score = 86.86
RF_score = 86.95
EXT_score = 85.82
KNN_score = 84.52
BAG_score = 87.01
PAST_score = 87.04
BST_score = 87.12
STCK_score = 87.23
SVM_score = 86.40
NN_score = 86.91
```

```python
scores = pd.Series([DT_score, RF_score, EXT_score, KNN_score,BAG_score,PAST_score, BST_score, STCK_score, SVM_score, NN_score], name='accuracy')
model = pd.Series(['Decision Tree', 'Random Forest', 'Extra Trees', 'K-Nearest Neighbor','Bagging','Pasting', 'Boosting','Stacking', 'Support Vector Machines', 'Reti neurali'], name='model')
```

```python
risultati = pd.concat([model,scores], axis=1)
risultati
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.set_xlim([82,89])
sns.barplot(data=risultati, y='model',x='accuracy', ax=ax)
ax.bar_label(ax.containers[0])

plt.savefig('risultati',dpi=400,bbox_inches="tight")
```
